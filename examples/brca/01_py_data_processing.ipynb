### merged_data 전처리

```python
df_merged = pd.read_csv("final_data/merged_data.txt", sep="\t")
# for group
df_cnv = pd.read_csv("final_data/cnv_annotation.txt", sep="\t")
df_methyl = pd.read_csv("final_data/methyl_annotation.txt", sep="\t")
df_mrna = pd.read_csv("final_data/mrna_annotation.txt", sep="\t")
df_mutation = pd.read_csv("final_data/mutation_annotation.txt", sep="\t")
```

```python
df_group = pd.concat((df_merged[["ID"]], df_merged[df_sample_selected["bcr_patient_barcode"]]), axis=1)
df_group
```

```python
df_group_cnv = pd.merge(df_group, df_cnv[["ID", "Gene"]], how="outer", on="ID")
df_group_cnv
```

```python
df_group_cnv_methly = pd.merge(df_group_cnv, df_methyl[["ID", "Gene_Symbol"]], how="outer", on="ID")
df_group_cnv_methly
```

```python
df_group_cnv_methly_mutation = pd.merge(df_group_cnv_methly, df_mutation[["ID", "gene_name"]], how="outer", on="ID")
df_group_cnv_methly_mutation
```

```python
df_group_cnv_methly_mutation_mrna = pd.merge(df_group_cnv_methly_mutation, df_mrna, how="outer", left_on="ID", right_on="x")
df_group_cnv_methly_mutation_mrna
```

```python
df_group["group"] = df_group_cnv_methly_mutation_mrna["Gene"].fillna(df_group_cnv_methly_mutation_mrna["Gene_Symbol"]).fillna(df_group_cnv_methly_mutation_mrna["gene_name"]).fillna(df_group_cnv_methly_mutation_mrna["x"])
df_group = df_group.dropna(subset=["group", "ID"])
df_group.index = range(df_group.shape[0])
df_group
```

```python
# select only mrna
df_group = df_group_cnv_methly_mutation_mrna.dropna(subset=["x", "ID"])
df_group = df_group.drop(columns=["Gene", "Gene_Symbol", "gene_name"])
df_group = df_group.rename(columns={"x": "group"})
df_group.index = range(df_group.shape[0])
df_group
```

```python
variable_to_index = dict()
group_to_index = dict()

k = 1
for i, row in df_group.iterrows():
    group = row["group"].split(";")
    for g in group:
        if g not in group_to_index.keys():
            group_to_index[g] = k
            k += 1

    variable = row["ID"]
    variable_to_index[variable] = i + 1
```

```python
from collections import defaultdict

variable_to_groups = defaultdict(list)
group_to_variables = defaultdict(list)
for i, row in df_group.iterrows():
    # one variable multiple groups
    # if ";" in row["group"]:
    group = row["group"].split(";")
    variable = row["ID"]
    for g in group:
        variable_to_groups[variable].append(g)
        group_to_variables[g].append([variable])
```

```python
import json

# root = "preprocessed"
root = "only_mrna"

with open(f"{root}/variable_to_index.json", "w") as json_file:
    json.dump(variable_to_index, json_file)

with open(f"{root}/index_to_variable.json", "w") as json_file:
    json.dump({value: key for key, value in variable_to_index.items()}, json_file)

with open(f"{root}/group_to_index.json", "w") as json_file:
    json.dump(group_to_index, json_file)

with open(f"{root}/index_to_group.json", "w") as json_file:
    json.dump({value: key for key, value in group_to_index.items()}, json_file)

with open(f"{root}/variable_to_groups.json", "w") as json_file:
    json.dump(variable_to_groups, json_file)

with open(f"{root}/group_to_variables.json", "w") as json_file:
    json.dump(group_to_variables, json_file)
```

```python
len(group_to_variables.keys())
```

```python
list_group_to_variables = [[]] * (len(group_to_variables.keys()) + 1)
for group in group_to_variables.keys():
    group_index = group_to_index[group]
    list_group_to_variables[group_index] = []
    for value in group_to_variables[group]:
        list_group_to_variables[group_index].append(variable_to_index[value[0]])
```

```python
import pickle

with open(f"{root}/group_to_variables_index.pkl", "wb") as f:
    pickle.dump(list_group_to_variables[1:], f)
```
